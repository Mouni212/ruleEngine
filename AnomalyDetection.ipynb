```python
import datetime
import os
import random
from adtk.visualization import plot
import pandas as pd
import csv
from adtk.detector import GeneralizedESDTestAD
from adtk.data import validate_series
```

```python
s = pd.read_csv('document.csv', index_col="created_date", parse_dates=True, squeeze=True)
s = validate_series(s)
esd_ad = GeneralizedESDTestAD(alpha=0.3)
anomalies = esd_ad.fit_detect(s)
plot(s, anomaly_pred=anomalies, ts_linewidth=2, ts_markersize=3, ap_markersize=5, ap_color='red',
     ap_marker_on_curve=True);
```
